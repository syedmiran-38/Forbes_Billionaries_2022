# **Forbes Billionaires 2022**

**Importing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

**Loading the DataSet**

```python
df = pd.read_csv('/content/forbes_billionaires_2022.csv')
df.head()
```

**Basic Analysis**

```python
df.shape
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

There are many NULL values

**Replacing Null values with appropriate values**

```python
df['title'].fillna('Unavailable', inplace = True)
```

```python
df.isnull().sum()
```

# **Answering Questions**

**Q1) Who are the top 10 richest CEOs ?**

```python
ceos = df[df['title'].str.contains('CEO')].head(10)
```

```python
ceos
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x = ceos['personName'], y = ceos['finalWorth'], hue = ceos['country'], dodge = False, palette = 'CMRmap')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('Top Richest CEOs', weight='bold').set_fontsize('30')
ax.set_ylabel("Final Worth",fontsize=20)
ax.set_xlabel("Name of Person",fontsize=20)
plt.show()
```

**Q2) What are the Top 10 cities in the world with most Richest Persons**

```python
top10 = pd.DataFrame(df['city'].value_counts().head(10))
```

```python
top10
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x = top10.index, y = top10['city'], dodge = False, palette = 'autumn')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('Top 10 cities in the world with most Richest Persons', weight='bold').set_fontsize('30')
ax.set_ylabel("No. of Persons",fontsize=20)
plt.show()
```

**Q3) Top 10 Countries with most Richest persons**

```python
top_10 = pd.DataFrame(df['country'].value_counts().head(10))
```

```python
top_10
```

**Q4) Top 10 Richest persons in India**

```python
ind = df[df['country'] == 'India'].head(10)
```

```python
ind.head(10)
```

```python
fig, ax = plt.subplots(figsize=(15, 10))
sns.barplot(x = ind['personName'], y = ind['finalWorth'], hue = ind['city'], dodge = False, palette = 'bright')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.title('Top 10 Richest persons in India', weight='bold').set_fontsize('30')
ax.set_xlabel("Final Worth",fontsize=20)
ax.set_ylabel("Name",fontsize=20)
plt.show()
```
